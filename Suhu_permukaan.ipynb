```python
#import library
import datetime
import pandas as pd
import matplotlib.pyplot as plt
#baca dataset
data = pd.read_csv('suhu_permukaan.csv')
```

```python
from matplotlib.path import WeakValueDictionary
import matplotlib.pyplot as plt
import numpy as np

#mengelompokkan data
time = data['datetime']
asimilasi = data['asimilasi']
observasi = data['observasi']
non_asimilasi = data['non asimilasi']

plt.figure(figsize=(13, 5))
plt.plot(time, asimilasi, label='Asimilasi',marker='*',zorder=2)
plt.plot(time, observasi, color='yellow', label='Observasi',marker='*',zorder=2)
plt.plot(time, non_asimilasi, color='red', label='Non Asimilasi',marker='*',zorder=2)

for temp in range(20, 28):
    plt.axhline(y=temp, color='gray', linestyle='-', linewidth=0.5,zorder=1)

plt.title('Suhu Permukaan', loc='center', pad=30, fontsize=20, color='black')
plt.xlabel('Waktu', fontsize=15)
plt.ylabel('Celcius', fontsize=15)
plt.grid(color='darkgray', linestyle=':', linewidth=0.5)
labels, locations = plt.yticks()
plt.legend(loc='right', bbox_to_anchor=(1.1, 1), shadow=True, ncol=1)
plt.tight_layout()
plt.show()
```

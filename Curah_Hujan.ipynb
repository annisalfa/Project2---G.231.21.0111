```python
#import library
import datetime
import pandas as pd
import matplotlib.pyplot as plt
#baca dataset
dataset = pd.read_csv('curah_hujan.csv')
```

```python
dataset
```

```python
import matplotlib.pyplot as plt
import numpy as np

#mengelompokan data menurut tahun
tahun = dataset['Bulan']
tahun_1981_1990 = dataset['1981-1990']
tahun_1991_2000 = dataset['1991-2000']
tahun_2001_2010 = dataset['2001-2010']

#membuat Multibar Chart
x = list(range(len(tahun)))  # Mengubah range menjadi list
width = 0.2

plt.bar(x, tahun_1981_1990, width, label='1981-1990',zorder=2)
plt.bar(np.add (x, width), tahun_1991_2000, width, label='1991-2000',zorder=2, color='red')
plt.bar(np.add (x, 2*width), tahun_2001_2010, width, label='2001-2010',zorder=2, color='#32CD32')
plt.xlabel('Bulan', fontsize=12)
plt.ylabel('Curah Hujan(mm)', fontsize=12)
plt.legend(loc='upper center', ncols=3, frameon = False, bbox_to_anchor=(0.5,1.1))
plt.ylim(0,700)
kelipatan = 20
batas = 700
for i in range(kelipatan, batas + 1, kelipatan):
    plt.axhline(y=i, color='gray', linestyle='-', linewidth=1,zorder=1)
plt.xticks(x,tahun)
plt.tight_layout()
plt.show()
```
